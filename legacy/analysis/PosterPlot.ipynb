```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img_cls = cv2.imread("/local_storage/datasets/sgerard/xview2/tier3/masks/joplin-tornado_00000000_post_disaster.png", cv2.IMREAD_GRAYSCALE)
img_loc = cv2.imread("/local_storage/datasets/sgerard/xview2/tier3/masks/joplin-tornado_00000000_pre_disaster.png", cv2.IMREAD_GRAYSCALE)
```

```python
img_loc_col = cv2.applyColorMap(img_loc, cv2.COLORMAP_VIRIDIS)
img_cls_col = cv2.applyColorMap(img_cls*int(255/4), cv2.COLORMAP_VIRIDIS)
cv2.imwrite("/tmp/loc.png", img_loc_col)
cv2.imwrite("/tmp/cls.png", img_cls_col)
```

```python
xs= [0.7549, 0.678, 0.7557, 0.7112, 0.726]
```

```python
plt.scatter(xs, ["Simplified"]*len(xs))
plt.xlim([0.0, 0.85])
plt.vlines(0.804, "Simplified", "Simplified")
```

```python
import matplotlib
FONTSIZE=48
matplotlib.rcParams.update({'font.size': FONTSIZE, "font.weight":"normal", "font.family":"Liberation Sans"})
fig = plt.figure(figsize=(15.5,6), dpi=300)
#plt.hlines(1,0,1, colors="black")  # Draw a horizontal line
plt.eventplot(xs, orientation='horizontal', colors='#3177c5', 
              label="Simplified model (varying seeds)", linewidth=8)
plt.vlines(0.804, 0.5, 1.5, colors="#D81B60", label="Winning solution", linewidth=8)
plt.xticks([0, 0.70, 0.8], rotation=45)
plt.yticks([])
plt.ylim(0.5, 1.5)
#fig.axes[0].spines['top'].set_visible(False)
#fig.axes[0].spines['right'].set_visible(False)
#fig.axes[0].spines['bottom'].set_visible(False)
#fig.axes[0].spines['left'].set_visible(False)
plt.xlabel('Competition score')
fig.axes[0].xaxis.set_label_coords(.38, -.1)
fig.axes[0].tick_params(which="major", length=20)
#plt.legend(loc="lower left")
#plt.axis('off')
plt.tight_layout()

plt.savefig("xview_poster_visualization.png", dpi=300)
plt.show()
```

```python
import matplotlib
FONTSIZE=32
matplotlib.rcParams.update({'font.size': FONTSIZE})
fig = plt.figure(figsize=(15,6), dpi=300)
plt.hlines(1,0,1, colors="black")  # Draw a horizontal line
plt.eventplot(xs, orientation='horizontal', colors='#1E88E5', label="Simplified model (varying seeds)")
plt.vlines(0.804, 0.5, 1.5, colors="#D81B60", label="Winning solution")
plt.xticks([0, 0.70, 0.8], rotation=90)
plt.yticks([])
plt.ylim(0.5, 1.5)
fig.axes[0].spines['top'].set_visible(False)
fig.axes[0].spines['right'].set_visible(False)
fig.axes[0].spines['bottom'].set_visible(False)
fig.axes[0].spines['left'].set_visible(False)
plt.xlabel('Competition score')
plt.legend(loc="upper left")
#plt.axis('off')
plt.tight_layout()
plt.savefig("xview_poster_visualizationv2.png", dpi=300)
plt.show()
```
